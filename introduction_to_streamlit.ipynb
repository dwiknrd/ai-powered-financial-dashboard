# <a id='toc2_'></a>[🚀 Introduction to Streamlit](#toc0_)

**🤔 Challenges for Data Scientist**

Sebagai seorang data scientist, tantangannya tidak hanya terbatas pada peran sebagai penganalisis data, tetapi juga sebagai penyedia aplikasi yang mempermudah orang lain untuk mendapatkan manfaat dari analisis data tersebut. Salah satu contohnya adalah kebutuhan akan web app yang mudah diakses yang menampilkan dashboard analytics atau fitur interaktif lainnya.

Dengan demikian, kita membutuhkan alat yang lebih dari sekadar melakukan analisis melalui Jupyter Notebook, tetapi juga mampu menghasilkan aplikasi web interaktif yang memungkinkan pengguna untuk berinteraksi langsung dengan data secara visual dan intuitif.

**💡 Solution**

Salah satu solusinya adalah menggunakan beberapa tools seperti ***Flask, Django, Dash,*** dan sebagainya. Namun, beberapa tools tersebut terlalu kompleks dan memerlukan pemahaman tambahan seperti HTML, CSS, dan JavaScript, serta proses deployment yang rumit.

Streamlit memungkinkan kita untuk dengan cepat membangun aplikasi web interaktif <u>**hanya dengan menggunakan bahasa Python**</u>, tanpa perlu keahlian tambahan dalam pengembangan web atau konfigurasi deployment yang rumit.

## <a id='toc2_1_'></a>[What is Streamlit](#toc0_)

Streamlit adalah sebuah framework Python yang digunakan untuk membuat aplikasi web interaktif dengan mudah. Dengan Streamlit, pengguna dapat mengembangkan aplikasi web tanpa memerlukan pengetahuan mendalam tentang pengembangan web tradisional.

## <a id='toc2_2_'></a>[Why Streamlit](#toc0_)

Alasan menggunakan Streamlit termasuk:

- Memenuhi kebutuhan analisis data yang kompleks.
- Mendukung keputusan berbasis data yang akurat.
- Proses pengembangan yang efisien dan cepat.
- Implementasi model machine learning dengan mudah dan efektif.

## <a id='toc2_3_'></a>[Advantages](#toc0_)

- Memungkinkan pengembangan aplikasi dengan cepat dan interaktif.
- Menggunakan bahasa Python secara eksklusif tanpa perlu mempelajari teknologi web lainnya.
- Menerima input dari pengguna dengan mudah dan menyediakan respons langsung.
- Proses deployment aplikasi yang cepat dan sederhana.
- Ramah pengguna, cocok untuk pengembang dengan berbagai tingkat keahlian.

## <a id='toc4_1_'></a>[✍️ Write](#toc0_)

### <a id='toc4_1_1_'></a>[`st.write()`](#toc0_)

[`st.write()`](https://docs.streamlit.io/develop/api-reference/write-magic/st.write) adalah salah satu fungsi paling serbaguna dalam Streamlit. Fungsi ini dapat digunakan untuk menampilkan berbagai jenis data seperti **teks, angka, DataFrame, gambar, grafik Matplotlib, dan banyak lagi.** Fungsi ini secara otomatis akan mendeteksi jenis data yang kita berikan dan menampilkannya dengan cara yang sesuai.

Berikut adalah beberapa contoh penggunaan `st.write()` untuk menampilkan berbagai jenis data:

<iframe

></iframe>

1. **Menampilkan Teks**

```python
import streamlit as st

st.write("Ini adalah teks sederhana.")
st.write("Anda bisa menggunakan *Markdown* untuk format teks.")
```

2. **Menampilkan Angka**

```python
number = 42
st.write("Angka favorit saya adalah", number)
```

3. **Menampilkan DataFrame**

```python
import pandas as pd

data = {
    'Kolom A': [1, 2, 3],
    'Kolom B': [4, 5, 6]
}
df = pd.DataFrame(data)
st.write("Ini adalah DataFrame:", df)
```

4. **Menampilkan Grafik Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np

# Buat data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Buat grafik
fig, ax = plt.subplots()
ax.plot(x, y)

# Tampilkan grafik dengan st.write()
st.write(fig)
```

5. **Menampilkan Gambar**

```python
# Menampilkan gambar menggunakan st.write() dan markdown
st.write("""
![Alt Text](https://www.example.com/path/to/your/image.jpg)
""")
```

6. **Menampilkan Objek Python Lainnya**

```python
# Menampilkan dictionary
dict_data = {"key1": "value1", "key2": "value2"}
st.write("Ini adalah dictionary:", dict_data)

# Menampilkan list
list_data = ["item1", "item2", "item3"]
st.write("Ini adalah list:", list_data)
```

## <a id='toc4_3_'></a>[📄 Data Elements](#toc0_)

### <a id='toc4_3_1_'></a>[`st.dataframe`](#toc0_)

`st.dataframe` digunakan untuk menampilkan data dalam format tabel interaktif. Pengguna dapat melakukan berbagai interaksi dengan tabel ini, seperti menggulir, mengurutkan, dan memfilter data. Fungsi ini sangat berguna ketika kita ingin menampilkan data yang dapat diakses dan dianalisis oleh pengguna.

## <a id='toc4_4_'></a>[📐 Layouts and Containers](#toc0_)

### <a id='toc4_4_1_'></a>[`st.columns`](#toc0_)

`st.columns` digunakan untuk membagi layar aplikasi menjadi beberapa kolom. Ini sangat berguna untuk mengatur tata letak elemen-elemen aplikasi kita.

**Contoh Penggunaan:**

```python
import streamlit as st

# Memisahkan layar menjadi 2 kolom
col1, col2 = st.columns(2)

# Menampilkan teks di kolom pertama
with col1:
    st.header("Kolom Pertama")
    st.text("Ini adalah teks di kolom pertama.")

# Menampilkan teks di kolom kedua
with col2:
    st.header("Kolom Kedua")
    st.text("Ini adalah teks di kolom kedua.")
```

### <a id='toc4_4_5_'></a>[`st.tabs`](#toc0_)

`st.tabs` digunakan untuk menampilkan elemen-elemen dalam tab-tab yang dapat digunakan pengguna untuk mengakses konten yang berbeda.

**Contoh Penggunaan:**

```python
import streamlit as st

# Menampilkan konten dalam tab-tab
tabs = st.tabs(["Tab 1", "Tab 2", "Tab 3"])

# Konten Tab 1
with tabs[0]:
    st.header("Konten Tab 1")
    st.text("Ini adalah konten dari Tab 1.")

# Konten Tab 2
with tabs[1]:
    st.header("Konten Tab 2")
    st.text("Ini adalah konten dari Tab 2.")

# Konten Tab 3
with tabs[2]:
    st.header("Konten Tab 3")
    st.text("Ini adalah konten dari Tab 3.")
```

- [**st.slider**](https://docs.streamlit.io/develop/api-reference/widgets/st.slider)
  - Menambahkan slider untuk memilih nilai numerik dalam rentang yang ditentukan.

```python

import streamlit as st

age = st.slider("How old are you?", 0, 130, 25)
st.write("I'm ", age, "years old")

values = st.slider(
    "Select a range of values",
    0.0, 100.0, (25.0, 75.0))
st.write("Values:", values)

from datetime import time

appointment = st.slider(
    "Schedule your appointment:",
    value=(time(11, 30), time(12, 45)))
st.write("You're scheduled for:", appointment)

start_time = st.slider(
    "When do you start?",
    value=datetime(2020, 1, 1, 9, 30),
    format="MM/DD/YY - hh:mm")
st.write("Start time:", start_time)

```

### <a id='toc4_5_4_'></a>[Elemen Input Lainnya](#toc0_)
- [**st.file_uploader**](https://docs.streamlit.io/develop/api-reference/widgets/st.file_uploader)
  - Menambahkan elemen untuk mengunggah file dari komputer pengguna.

```python
import streamlit as st

uploaded_files = st.file_uploader("Choose a CSV file", accept_multiple_files=True)
for uploaded_file in uploaded_files:
    bytes_data = uploaded_file.read()
    st.write("filename:", uploaded_file.name)
    st.write(bytes_data)
```

# <a id='toc5_'></a>[⚙️ Konfigurasi](#toc0_)

Streamlit menyediakan fitur konfigurasi yang membantu kita dalam menyesuaikan perilaku dan tampilan aplikasi web yang kita buat. Salah satu fungsi konfigurasi yang berguna adalah:

## <a id='toc5_1_'></a>[`st.set_page_config`](#toc0_)

[`st.set_page_config`](https://docs.streamlit.io/develop/api-reference/configuration/st.set_page_config) digunakan untuk menentukan konfigurasi halaman aplikasi Streamlit. Fungsi ini memungkinkan kita untuk mengatur berbagai parameter, termasuk:

1. **`page_title`**
   - **Deskripsi**: Judul halaman yang akan ditampilkan di tab browser.
   - **Tipe**: String
   - **Contoh**: `st.set_page_config(page_title="My Streamlit App")`

2. **`page_icon`**
   - **Deskripsi**: URL ikon favicon untuk ditampilkan di tab browser.
   - **Tipe**: String (URL)
   - **Contoh**: `st.set_page_config(page_icon="🎈")`

3. **`layout`**
   - **Deskripsi**: Tipe layout halaman ("centered" atau "wide").
   - **Tipe**: String
   - **Contoh**: `st.set_page_config(layout="wide")`

4. **`initial_sidebar_state`**
   - **Deskripsi**: Kondisi awal sidebar ("auto", "expanded", atau "collapsed").
   - **Tipe**: String
   - **Contoh**: `st.set_page_config(initial_sidebar_state="expanded")`

5. **`menu_items`**
   - **Deskripsi**: Daftar item-menu di sidebar aplikasi.
   - **Tipe**: Dictionary
   - **Contoh**: `st.set_page_config(menu_items={"About": "#", "Help": "#"})`


# <a id='toc6_'></a>🔗 [Connect Streamlit to Data Source](https://docs.streamlit.io/develop/tutorials/databases) [&#8593;](#toc0_)

Selain bisa membuka data secara lokal, Streamlit juga bisa digunakan untuk berkoneksi dengan sumber data lainnya seperti database SQL, API web, dan Google Sheets.

Pada bagian ini kita tidak akan membahas semuanya, kita akan fokus pada cara menghubungkan dashboard dengan sumber data Google Sheets. Ini penting karena kita ingin menyediakan dashboard yang bisa menampilkan hasil analisis data secara dinamis dari Google Sheets.

## <a id='toc6_1_'></a>[Connect Streamlit to a public Google Sheet](https://docs.streamlit.io/develop/tutorials/databases/public-gsheet) [&#8593;](#toc0_)

Google Sheet merupakan salah satu sumber data yang dinamis. Dinamis yang dimaksud adalah data dapat diubah dan diperbarui secara real-time oleh beberapa pengguna. Sehingga apabila kita ingin menampilkan hasil analisis dari data yang dinamis, kita dapat menampilkannya ke dalam dashboard Streamlit.

Kita akan mencoba untuk menghubungkan Google Sheet publik ke dalam dashboard Streamlit kita. Meskipun Google Sheet yang kita gunakan bersifat publik, kita dapat menyembunyikan link tersebut, sehingga pengguna tetap tidak dapat mengakses data secara langsung. Hal ini juga membantu menjaga privasi dan keamanan data.


Langkah-langkah yang perlu kita lakukan untuk menghubungkan data dengan dashboard kita adalah:

**1. Sediakan URL Google Sheet dari data yang akan digunakan**
   - Pastikan bahwa Google Sheet telah diatur menjadi **publik**.

**2. Tambahkan URL Sheet ke Local App Secret kita**
   - Pada project kita, tambahkan satu folder baru dengan nama `.streamlit`.
   - Di dalam folder tersebut, tambahkan file baru dengan nama `secrets.toml`.
   - Buka file `secrets.toml` dan tambahkan URL Sheet dengan format seperti contoh berikut:
     ```toml
     [connections.gsheets]
     spreadsheet = "https://docs.google.com/spreadsheets/d/xxxxxxx/edit#gid=0"
     ```

**3. Tulis Sintaks untuk Meload Data di App Streamlit**

```python
# Import library yang diperlukan
import streamlit as st
from streamlit_gsheets import GSheetsConnection

# Buat variabel koneksi
conn = st.connection("gsheets", type=GSheetsConnection)

# Baca data
df = conn.read()
```

Setelah kita membuat variabel `df`, maka kita sudah berhasil membaca data kita dari Google Sheet dan siap untuk ditampilkan atau dianalisis lebih lanjut di dashboard Streamlit kita.

### <a id='toc6_1_1_'></a>[Additional](#toc0_)

#### <a id='toc6_1_1_1_'></a>[`secrets.toml`](#toc0_)

File ini merupakan jenis file tersembunyi yang digunakan untuk menyimpan informasi sensitif, seperti URL Google Sheet atau kredensial API. File ini hanya digunakan di lokal dan tidak akan diupload ketika proses deployment. Hal ini menjaga keamanan informasi sensitif yang kita gunakan.

Dalam penulisannya, kita hanya perlu menuliskan judul serta suatu variabel yang nantinya akan kita panggil atau gunakan di file app kita. Berikut adalah formatnya:

```toml
[judul]
variabel1 = "...."

[judul_2]
variabel2 = "..."
```

Contoh penggunaan untuk menghubungkan Google Sheet:
```toml
[gsheets]
spreadsheet = "https://docs.google.com/spreadsheets/d/xxxxxxx/edit#gid=0"
```

File `secrets.toml` ini sangat penting untuk menjaga keamanan dan privasi informasi yang kita gunakan dalam aplikasi Streamlit, dan memastikan bahwa informasi sensitif tersebut tidak terlihat oleh orang lain ketika aplikasi kita dideploy ke server atau dibagikan ke orang lain.

#### <a id='toc6_1_1_2_'></a>[`gsheets-connection`](https://github.com/streamlit/gsheets-connection) [&#8593;](#toc0_)


Library ini digunakan untuk menghubungkan aplikasi Streamlit dengan Google Sheets, memungkinkan kita untuk membaca dan mengelola data secara dinamis. Pada pengaturannya, kita menggunakan fitur dari Streamlit yaitu `st.connection` dan memasukkan parameter `type=GSheetsConnection`. Kode tersebut akan menghasilkan koneksi yang selanjutnya dapat kita chaining dengan fungsi `read()`.

Di dalam fungsi `read()`, terdapat beberapa parameter yang bisa kita atur di antaranya:
- `spreadsheet`: URL atau ID dari Google Sheet yang akan dibaca.
- `worksheet`: Nama atau indeks worksheet yang akan dibaca.
- `ttl`: Time-to-live cache dalam detik, untuk mengatur seberapa lama data yang diambil dari Google Sheet disimpan dalam cache.
- `usecols`: Kolom yang akan dibaca dari Google Sheet.
- `nrows`: Jumlah baris yang akan dibaca dari Google Sheet.

Berikut adalah contoh penggunaannya:
```python
df = conn.read(
    spreadsheet=st.secrets.gsheets["spreadsheet"],
    worksheet="773835363",
    ttl="10m",
    usecols=[0, 1],
    nrows=3,
)
```

# <a id='toc7_'></a>[🌐 Deploy](#toc0_)

Setelah kita membuat dashboard dengan fitur yang sesuai dengan kebutuhan, tentu kita ingin dashboard tersebut bisa diakses secara umum. Artinya, web dashboard kita tidak lagi berjalan di lokal kita, namun di server cloud. Proses pengiriman aplikasi dari lokal ke produksi atau live dinamakan deployment.

## <a id='toc7_1_'></a>[Deployment Concept](#toc0_)

Pada dasarnya, proses deploy ini adalah kita memindahkan semua kebutuhan untuk menampilkan aplikasi kita dan menjalankannya di suatu server secara terus menerus. Sehingga, kita perlu menyediakan lingkungan server yang bisa menjalankan aplikasi kita tersebut dengan baik. Beberapa hal yang perlu dipersiapkan antara lain:

1. **Server atau Cloud Provider**: Penyedia layanan yang akan menjalankan aplikasi kita, seperti Heroku, AWS, Google Cloud, atau Streamlit Cloud.
2. **Dependencies**: Semua library dan paket yang digunakan dalam aplikasi harus diinstal di server.
3. **Configuration Files**: File konfigurasi seperti `requirements.txt` yang menjelaskan dependencies dan pengaturan lainnya.
4. **Environment Variables**: Variabel lingkungan yang digunakan untuk menyimpan informasi sensitif seperti API keys, database URLs, dan lain-lain.
5. **Version Control**: Menggunakan sistem kontrol versi seperti Git untuk melacak perubahan dan memudahkan proses deployment.

## <a id='toc7_2_'></a>[Deployment Methods](https://docs.streamlit.io/deploy/tutorials) [&#8593;](#toc0_)

Ada beberapa metode untuk melakukan deployment aplikasi Streamlit:

- **Streamlit Community Cloud**: Layanan gratis yang disediakan oleh Streamlit untuk hosting aplikasi Streamlit.
- **Heroku**: Platform cloud yang mendukung berbagai bahasa pemrograman dan framework, termasuk Streamlit.
- **Docker**: Containerization yang memungkinkan aplikasi berjalan dalam lingkungan yang terisolasi.
- **AWS, Azure, dan GCP**: Penyedia cloud besar yang mendukung deployment aplikasi Streamlit dengan fleksibilitas dan skalabilitas tinggi.

## <a id='toc7_3_'></a>Contoh [Deployment Streamlit di Streamlit Community Cloud](https://docs.streamlit.io/deploy/streamlit-community-cloud/deploy-your-app) [&#8593;](#toc0_)

Berikut adalah langkah-langkah untuk melakukan deployment aplikasi Streamlit di Streamlit Community Cloud:

1. **Push Kode ke GitHub**:
   - Buat repository baru di GitHub dan push kode aplikasi Streamlit kita ke repository tersebut.
```
your-repository/
├── your_app.py
└── requirements.txt
```
   - Jangan push file secret apapun, seperti: `secret.toml`, `.env`

2. **Login ke Streamlit Community Cloud**:
   - Buka [Streamlit Community Cloud](https://share.streamlit.io/) dan login menggunakan akun GitHub kita.

3. **Create App**:
   - Klik "Create app"
   - Pilih menu deploy dari Github repo
   - Tips: Gunakan menu *"Paste Github URL"* dan tempel URL github file python kita 

4. **Advance Setting**: Pilih bagian advance setting
   - Pilih python versi yang sesuai dengan versi python dilokal kita
   - Masukkan semua teks yang ada di file `secret.toml` dibagian secret

5. **Deploy Aplikasi**:
   - Sesuaikan URL app yang kita inginkan
   - Klik "Deploy!"

6. **Monitor dan Update**:
   - Setelah deployment selesai, kita dapat mengakses aplikasi kita melalui URL yang disediakan.
   - Untuk melakukan update, cukup push perubahan ke repository GitHub kita dan aplikasi akan otomatis di-deploy ulang.
